```python
from scipy.optimize import minimize

# Uma função qualquer (ex: quadrática)
def funcao_objetivo(x):
    return x[0]**2 - 4*x[0] + x[1]**2 - x[1] - x[0]*x[1]

# Ponto inicial
x0 = [10000.0, 10000.0]

# Rodando o Nelder-Mead
res = minimize(funcao_objetivo, x0, method='nelder-mead')

print(f"Mínimo encontrado em: {res.x}")
```

```python
def funcao_objetivo(x):
    return  (x[0]-10)**2+(x[1]+10)**2+(x[2]-2)**2

# Ponto inicial
x0 = [10000.0, 10000.0, 10000.0]

# Rodando o Nelder-Mead
res = minimize(funcao_objetivo, x0, method='nelder-mead')

print(f"Mínimo encontrado em: {res.x}")
```
